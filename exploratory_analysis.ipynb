```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict
import nltk
%matplotlib inline

nltk.download("punkt")
```

```python
# doesn't work without specifying engine (why???)
lyrics = pd.read_csv("data/taylor_swift_lyrics.csv", engine = "python")
```

```python
# key: word
# value: frequency
word_freqs = defaultdict(lambda: 0)

for i, row in lyrics.iterrows():
#     words_list = row["lyric"].split()
    words_list = nltk.word_tokenize(row["lyric"])
    for word in words_list:
        word_freqs[word] += 1
```
